# Double Link List Palindrome

```python
class Node():
    def __init__(self, value):
        self.info = value
        self.forward = None
        self.backward = None


class doubleLinkList:
    def __init__(self):
        self.start = None
        self.last = None
        self.count = 0

    def insert(self, value):
        node = Node(value)
        if self.start == None:
            self.start = node
            self.last = node
        else:
            self.last.foward = node
            node.backward = self.last
            self.last = node
        self.count+=1

    def display(self):
        p = self.start
        #print(p.foward.info, p.foward.foward.info, self.last.info)
        while p != self.last:
            print(p.info)
            p = p.foward
        print(p.info)
            


    def palindrome(self):
        p = self.start
        q = self.last
        middle = (self.count // 2)
        count = 0
        while count < middle:
            if p.info == q.info:
                p = p.foward
                q = q.backward
            else:
                return False
            count +=1
        return True
```

```python
class stacks():
    def __init__(self,size):
        self.items = [None] * (size+3)
        for i in range(0, len(self.items), len(self.items)//3):   
            print(i)
            if i != 1:
                print('got here')
                self.items[i-1] = '*'
                
    def push(self,value,stack):
        listStack = self.items
        perStack = len(self.items) // 3
        for i in range(int(perStack * (stack - 1)), perStack * stack):
            if listStack[i] == None:
                listStack[i] = value
                return
        raise Exception ('Stack is full')
        
        
    def pop(self,stack):
        listStack = self.items
        perStack = len(self.items) // 3
        prev = '*'
        for i in range((perStack * (stack-1)), perStack * stack):
            if prev != '*' and  listStack[i] == None:
                listStack[prevIndex] = None
                return
            prev = listStack[i]
            prevIndex = i
        raise Exception ('Stack is Empty')
            
    def peek(self,stack):
        listStack = self.items
        perStack = len(self.items) // 3
        prev = '*'
        for i in range((perStack * (stack-1)), perStack * stack):
            if prev != '*' and  listStack[i] == None:
                return listStack[prevIndex]
            prev = listStack[i]
            prevIndex = i
        return 'None'
    
    def display(self):
        print(self.items)
        
```

```python
class SetOfStacks():
    def __init__(self, maxPerStack):
        self.item = [[]]
        self.maxPerStack = maxPerStack
        self.count = 0
        self.currentStack = 0

    def pop(self):
        currentStack = int((self.count - 1) / self.maxPerStack)
        self.item[currentStack].pop()
        currentStack -=1
        
    def push(self, value):
        if self.count == 0:
            currentStacks = 0
        else:
             currentStacks = int((self.count) / self.maxPerStack)
                
        if self.currentStack < currentStacks:
            #stack is full
            self.item.append([])
            self.item[currentStacks].append(value)
            self.count +=1
            self.currentStack +=1
        else:
            #we are still in the same stack cuz its not full yet
            self.item[currentStacks].append(value)
            self.count +=1
        
            
    def  popAtIndex(self, index):
        if index > self.currentStack :
            raise Exception ('Stack does not exist')
        else:
            self.item[index].pop()
            
    def display(self):
        print(self.item)
```

```python
class graph():
    
    def __init__(self):
        self.root = {}
        
    def insert(self,start,end = ''):
        _graph = self.root
        if start in _graph:
            _graph[start].append(end)
        else:
            if end == '':
                _graph[start] = []
            else:
                _graph[start] = [end]
            
    def findDirection(self,start,end):
        _graph = self.root
        
        if start == end:
            return True
        elif start not in _graph:
            return
        else:
            for item in _graph[start]:
                if item != start:
                    value = self.findDirection(item, end)
                    if value == True:
                        return True
        return False
        
    def display(self):
        print(self.root)
        
            
    
            
```

# PROJECT AND DEPENDENCY GRAPH

```python
def buildOrder(projects, dependencies):
    #adjacencyMatrix = {a:[d], b:[d], c:[], d:[c], e:[], f:[a ,b] }
    order = []
    
    #if dependency is needed but is not in project:
    for dependency in dependencies:
        if dependency[0] not in projects:
            raise Exception (no valid build order)
            
            
    for project in projects:
        _buildOrder(project, dependencies, order)
    return order

def _buildOrder(project, dependencies, order):
    for dependency in dependencies:
        if dependency[0] == project and dependency[1] == project:
            raise Exception('Project cannot depend on itself')
            
        elif dependency[1] == project and dependency[0] not in order:
            _buildOrder(dependency[0], dependencies, order)
            if dependency[0] not in order:
                order.append(dependency[0])
            
    if project not in order:
        order.append(project)
    return order
```

# DEPTH FIRST SEARCH AND BREADTH FIRST SEARCH

```python
from collections import deque

class Graph():
    def __init__(self):
        self.root = {}
        
    def insert(self, start, end = ''):
        p = self.root
        if self.root == None:
            if end == '':
                p[start] = []
            else:
                p[start] = [end]
        else:
            if start in p.keys():
                p[start].append(end)
            else:
                if end == '':
                    p[start] = []
                else:
                    p[start] = [end]
                    
    def depthFirstSearch(self, start):
        stack = []
        result = []
        stack.append(start)
        return self._depthFirstSearch(start,result,stack)
        
    def _depthFirstSearch(self,start,result,stack):
        p = self.root
        while stack:
            if start not in result: 
                result.append(start)
                
            if p[start] == []:
                start = stack.pop()
                self._depthFirstSearch(start,result,stack)
                
            if len(stack) == 0:
                return result
            else:    
                for vertice in p[start]:
                    if vertice not in result and vertice not in stack:
                        if start not in stack:
                            stack.append(start) 
                        self._depthFirstSearch(vertice,result,stack)
                        break
                if len(stack) != 0:
                    vertice = stack.pop() 
                    self._depthFirstSearch(vertice,result,stack)
        return result
                
        
        
    def breadthFirstSearch(self, start):
        que = deque()
        result = []
        que.append(start)
        return self._breadthFirstSearch(start,que,result)
        
    def _breadthFirstSearch(self,start,que,result):
        p = self.root
        
        for vertices in p[start]:
            if vertices not in que and vertices not in result:
                que.append(vertices)
        if que[0] not in result:
            result.append(que.popleft())
            
        if len(que) != 0:
            self._breadthFirstSearch(que[0],que,result)
        else:
            return result
        
        
        return result
            
        
        
        
            
```

```python
def levelOrderTranversal(self):
    p = self.root
    tranversal = []
    que = deque()
    que.append(p.info)

    count = 0
    tranversal[count] = [p.info]
    return self._levelOrderTranversal(p, que, count)

def _levelOrderTranversal(self, p, que, count):

    while len(que) != 0:
        tranversal[count] = []
        count +=1
        temp = que.popleft()
        tranversal[count].append([[temp.lchild, temp.rchild],[ temp.rchild, temp.lchild]])
        self._levelOrderTranversal(temp.lchild, que, count)
        self._levelOrderTranversal(temp.rchild, que, count)

    return tranversal
```

# CHECK IF TREE IS A SUBTREE

```python
class Node():
    def __init__(self, value):
        self.info = value
        self.lchild = None
        self.rchild = None
        self.parent = None


class BinarySearchTree():
    def __init__(self):
        self.root = None

    def insert(self,value):
        p = self.root
        que = deque()
        que.append(p)
        
        
        if p == None:
            self.root = Node(value)
        else:
    
            while len(que) != 0:
                temp = que.popleft()

                if temp.lchild == None and temp.rchild == None:
                    temp.lchild = Node(value)
                    temp.lchild.parent = temp
                    break
                elif temp.lchild == None and temp.rchild != None:
                    temp.lchild = Node(value)
                    temp.lchild.parent = temp
                    break
                elif temp.lchild != None and temp.rchild == None:
                    temp.rchild = Node(value)
                    temp.rchild.parent = temp
                    break
                elif temp.lchild != None and temp.rchild != None:
                    que.append(temp.lchild)
                    que.append(temp.rchild)
                    
    
    def checkSubTree(self,T2):
        check = T2.root.info
        p = self.root
        que = deque()
        que.append(p)

        temp = None

        if p.info == check:
            temp = p

        while len(que) != None:
            temp = que.pop()

            if temp.lchild.info == check:
                temp = temp.lchild
                break

            if temp.rchild.info == check:
                temp = temp.rchild
                break

            if temp.lchild != None:
                que.append(temp.lchild)

            if temp.rchild != None:
                que.append(temp.rchild)


        if temp == None:
            raise Exception('Trees is not a subtree')
        else:
            _list1 = heightTranversal(temp)
            print(_list1)
            _list2 = heightTranversal(T2.root)
            print(_list2)
    
            if _list1 == _list2:
                print('Tree is a subtree')
            else:
                print('Tree is not a subtree')

                
                
def  heightTranversal(tree):
    p = tree
    que = deque()
    que.append(p)
    result = []

    while que:
        temp = que.pop()
        result.append(temp.info)
        
        if temp.lchild != None:
            que.append(temp.lchild)
        if temp.rchild != None:
            que.append(temp.rchild)
    return result
```

```python
class Node():
    def __init__(self, value, parent):
        self.info = value
        self.parent = parent
        self.lchild = None
        self.rchild = None
        

class tree():
    
    def __init__(self):
        self.root = None
        self.graph = {}

    def insert(self, value, parent = None):
        p = self.root
        que = deque()
        que.append(p)

        q = self.graph
        if parent not in q:
            q[parent] = [value]
        else:
            q[parent].append(value)
            
        
        while que:
            temp = que.popleft()
            
            if temp == None :
                self.root = Node(value, None)
                break

            if temp.lchild != None:
                que.append(temp.lchild)
            elif temp.lchild == None:
                temp.lchild = Node(value, parent)
                break

            if temp.rchild != None:
                que.append(temp.rchild)
            elif temp.rchild == None:
                temp.rchild = Node(value, parent)
                break


    def delete(self, value):
        p = self.root
        que = deque()
        que.append(p)
        prev = deque()
    
        #prev = None

        while que:
            temp = que.popleft()

            if temp.parent == None and temp.info == value:
                self.root = None
                break

            elif temp.info == value:
                if temp.lchild == None and temp.rchild == None:
                    print('got here', prev.info)
                    if prev.lchild.info == value:
                        prev.lchild = None
                        break
                    elif prev.rchild == value:
                        prev.rchild = None
                        break
                    
                elif temp.lchild == None and temp.rchild != None:
                    temp = temp.rchild
                    break
                    
                else:
                    deletedNode = temp
                    while temp.lchild != None:
                        prev = temp
                        #prev.append(temp)
                        temp = temp.lchild
                    deletedNode.info = temp.info
                    
                    if temp.rchild != None:
                        prev.lchild = temp.rchild
                        #prev.popleft().lchild = temp.rchild
                    else:
                        prev.lchild = None
                        #prev.popleft().lchild = None
                        
                    deletedNode.lchild = None
                    break
            else:
                
                
                if temp.lchild != None:
                    prev = temp
                    #prev.append(temp)
                    que.append(temp.lchild)
                if temp.rchild != None:
                    prev = temp
                    #prev.append(temp)
                    que.append(temp.rchild)
                print(prev.info)
    
    def height_traversal(self):
        result = []
        que = deque()
        p = self.root
        que.append(p)
        
        while que:
            temp = que.popleft()
            result.append(temp.info)
            
            if temp.lchild != None:
                que.append(temp.lchild)
            if temp.rchild != None:
                que.append(temp.rchild)
        
        return result
    
    
    def randomNode(self):
        result = self.height_traversal()
    
        return print(self.graph)#random.choice(result)
    
    
    def BFS(self,start):
        p = self.graph
        que = deque()
        que.append(start)
        result = []
        
        while que:
            temp = que.popleft()
            
            if temp in list(p.keys()):
                for neigbours in p[temp]:
                    que.append(neigbours)
            result.append(temp)
        return result

    
    
    def dFS(self,start):
        p = self.graph
        result = []
        stack = []
        stack.append(start)
        
        while stack:
            current = stack.pop()
            if current in list(p.keys()):
                for vertices in p[current]:
                    if vertices not in stack and vertices not in result:
                        stack.append(vertices)
            result.append(current)
        return result
    
    
    def pathSum(self, start, target):
        _list = self.BFS(start)
        self._anyequalto(_list,10)
        self.combination_in_range(10, 0, _list, B=[])
        
        
    def combination_in_range(self,V, t, A, B=[]):
        for i,a in enumerate(A):
            print(i)
            if a > V+t:    # B+[a] is too large
                continue

            # B+[a] can still be a possible list
            B.append(a)

            if a >= V-t:   # Found a set that works
                print(B)

            # recursively try with a reduced V
            # and a shortened list A
            self.combination_in_range(V-a, t, A[i+1:], B)

            B.pop()        # drop [a] from possible list

                 
                
                
    
    def _anyequalto(self,num_lst, n):
        """Return True if 2 numbers from `num_lst` add up to n, False otherwise."""
        num_set = set(num_lst)
        for i in num_set:
            if n - i in num_set:
                return True
        return False
    
    
    def anyequalto(self,num_lst, n):
        """Return True if 2 numbers from `num_lst` add up to n, False otherwise."""
        num_set = set(num_lst)
        return any(n - i in num_set for i in num_set)
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
    
```
